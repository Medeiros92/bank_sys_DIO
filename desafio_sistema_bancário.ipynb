```python
menu = """
    OPERAÇÕES:

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

saldo = 0
lim_saque = 3
val_max_saque = 500
extrato = ""
num_saque = 0

while True:
    
    opcao = input(menu)
    
    if opcao == "d":
        valor = float(input("Informe o valor depositado: "))
        
        if valor > 0:
            saldo += valor
            extrato += f"Depósito: R$ {valor:.2f}\n"
            
        else:
            print("***Operação falhou! O valor informado é inválido.***")
            
    elif opcao == "s":
        
        if num_saque <= lim_saque:
            num_saque +=1
            valor = float(input("Informe o valor do saque: "))
                
            if valor > 0:
                
                if valor <= val_max_saque:
                    
                    if saldo > valor:
                        saldo -= valor
                        extrato += f"Saque: R$ {valor:.2f}\n"

                    else:
                        print("***Operação falhou! Você não tem saldo suficiente.***")
                else:
                    print(f"***Operação falhou! O limite de saque é de R$ {val_max_saque}***")
            else:
                print("***Operação falhou! O valor informado é inválido.***")
        else:
            print("***Operação falhou! Você atingiu o número maximo de saques diários.***")
            
    elif opcao =="e":
        if not extrato:
            print("\n***Nenhum movimento foi registrado.***")
            
        else:
            print(f"""
========== Extrato ==========

{extrato}

=============================""")
            
    elif opcao == "q":
        break
        
    else:
        print("***Opção Inválida.***")
            
    

```
